```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# Load Dataset

iris = load_iris(as_frame=True)
df = iris.frame
```

```python
df.head()
```

```python
# Features And Target

x = iris.data
y = iris.target 

# Features Name
print("Features Name = ",iris.feature_names)
print("Target Class = ", iris.target_names)
```

```python
# Train - Test - Split 
# Train = 30%
# Test = 70%
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=42)
print("Train Data x :",x_train.shape)
print("Train Data y :",y_train.shape)

print("Test Data x :",x_test.shape)
print("Test Data y :",y_test.shape) 
```

```python
# Create Random Forest Model

rf_model = RandomForestClassifier(
    n_estimators=100, # Number of Tree
    criterion="gini", # Split Quality 
    random_state=42
)

# Train The Model

rf_model.fit(x_train,y_train)
```

```python
# Make Predicrtion

y_pred = rf_model.predict(x_test)
print("Prediction",y_pred[:10])
print("Actual    ",y_test[:10])
```

```python
# Evalute Model
# Accuracy

accuracy = accuracy_score(y_test,y_pred)
print("Accuracy ",accuracy)

# Classification Report 
print("classification Report = \n", classification_report(y_test,y_pred))

# Confusion Matrix
print("Confusion Matrix = \n", confusion_matrix(y_test,y_pred))
```

```python
# Feature Importance

importances = rf_model.feature_importances_
features = iris.feature_names

# Create Features
features_df = pd.DataFrame({
    "Feature" : features,
    "Importance" : importances
}).sort_values(by="Importance",ascending =False)
print(features_df)

plt.barh(features_df["Feature"],features_df["Importance"])
plt.xlabel("Imporatnces")
plt.ylabel("Frature Importance")
plt.show()
```

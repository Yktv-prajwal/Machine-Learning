```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline  import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
# Load DataSet

df = pd.read_csv("Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Delete Column Customer

df = df.drop(["customerID"],axis=1)
```

```python
df["Churn"] = df["Churn"].map({"No" : 0 , "Yes" : 1})
df.head()
```

```python
# Change Datatype

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df.dtypes
```

```python
# Filling Null

df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())
```

```python
df.head()
```

```python
# Feature and Target

x = df.drop(["Churn"],axis=1)
y = df["Churn"]
```

```python
y.head()
```

```python
unique = df["PaymentMethod"].unique()
unique
```

```python
df.columns
```

```python
Categorical_col = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']

numerical_col = ["tenure",'MonthlyCharges', 'TotalCharges']

# Preprocessing

preprocesser = ColumnTransformer(
    transformers=[
                ("num", StandardScaler(),numerical_col),
                ("cat", OneHotEncoder(drop="first",sparse_output=False),Categorical_col)
                ])

```

```python
# Train , Test , Split
# Train 80%
# Test 20%

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,stratify=y,random_state=42)

print("Train Data x :",x_train.shape)
print("Train Data y :",y_train.shape)

print("Test Data x :",x_test.shape)
print("Test Data y :",y_test.shape)
```

```python
# Create a model Pipeline

model = Pipeline(steps=[
    ("preprocesser",preprocesser),
    ("classifier",LogisticRegression(max_iter=1000,random_state=42))
])

# Train Model

model.fit(x_train,y_train)
```

```python
# Make Prediction 
y_pred = model.predict(x_test)
print("Accuracy : \n" ,accuracy_score(y_test,y_pred))
print("Classification Report : \n", classification_report(y_test,y_pred))
print("confusion_matrix \n",confusion_matrix(y_test,y_pred))
```

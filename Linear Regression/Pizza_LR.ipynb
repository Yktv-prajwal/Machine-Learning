```python
import numpy as np
from sklearn.linear_model import LinearRegression
import  matplotlib.pyplot as plt
```

```python
# Creat a Dataset 

X = np.array([8,10,12]).reshape(-1,1) # Pizza Size
Y = np.array([10,13,16]) # Pizza Price
```

```python
model = LinearRegression()
model.fit(X ,Y)

# Slope (m)

m = model.coef_[0]
print(f"Slope = {m}")

# Intercept

b = model.intercept_
print(f"Intercept (b) = {b}")
```

```python
# Price Predect Size = 20 inch

X_new = np.array([20]).reshape(-1,1)
Y_pred = model.predict(X_new)
print(f"20 Inch Pizza Price = ${Y_pred[0]:.2f}")
```

```python
# Plot

plt.scatter(X,Y,color = "Blue", label = "Data Point")
plt.plot(X,model.predict(X),color = "red",label = "Regression Line")
plt.xlabel("inches")
plt.ylabel("Price $")
plt.title("Piza Price Prediction")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load Dataset
df = pd.read_excel("pizza_price_dataset.xlsx")
df.head()

# Feature and Target 
X = df[["Diameter_Inches"]]
Y = df[["Price_USD"]]

# Train Model

model = LinearRegression()
model.fit(X,Y)
```

```python
# Find Intercept And Slpoe

m = model.coef_[0]
b = model.intercept_
print(f"Slope m = {m}")
print(f"Intercept b = {b}")
```

```python
# Prediction

new_pizza = [[25]]
prediction = model.predict(new_pizza)
print(f"New 25 inch Pizza Price = {prediction[0]}")
```

```python
# Plot
plt.figure(figsize=(15,8))
plt.scatter(X,Y,color = "Blue", label = "Data Point")
plt.plot(X,model.predict(X),color = "red",label = "Regression Line")
plt.xlabel("inches")
plt.ylabel("Price $")
plt.title("Piza Price Prediction")
plt.legend()
plt.show()
```
